```python
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode(connected=True)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('2014_World_Power_Consumption.csv')
df.head()
```

```python
data = dict(
    type='choropleth',
    locations=df['Country'],            # Use the country names from your DataFrame
    locationmode='country names',
    z=df['Power Consumption KWH'],
    text=df['Country'],
    colorbar={'title': 'Power Consumption KWH'}
)
```

```python
layout = dict(title='2014 World Power Consumption',
              geo=dict(showframe=False, projection={'type': 'mercator'}))
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap,validate=False)
```

```python
usdf = pd.read_csv('us_county_level_presidential_results_12_16.csv')
usdf.head()
```

```python
data = dict(type='choropleth',
            colorscale='Viridis',
            reversescale=True,
            locations=usdf['combined_fips'],  # Spatial coordinates
            z=usdf['per_dem_2016'],  # Data to be color-coded
            locationmode='USA-states',  # set of locations match entries in `locations`
            text=usdf['votes_dem_2016'],
            colorbar={'title': '% Democratic votes'})
```

```python
layout = dict(title='2011 US Agriculture Exports by State',
              geo=dict(showframe=False, projection={'type': 'natural earth'}))
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap, validate=False)
```
