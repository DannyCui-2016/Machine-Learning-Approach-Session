```python

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', figsize=(8,6), c='red')
```

```python
iris['sepal_length'].hist(bins=50, alpha=0.20)
plt.style.use('ggplot')
plt.savefig('histogram.png')
```

```python
iris['sepal_length'].hist(bins=50, alpha=0.4, color='darkgreen')
plt.style.use('ggplot')
plt.savefig('histogram2.png')
```

```python
iris['sepal_width'].plot.box(color='orange')
plt.savefig('boxplot..png')
```

```python
iris['sepal_length'].plot.kde(lw=1, linestyle='--', color='red')
plt.title('KDE Plot of Sepal Length')
plt.xlabel('Sepal Length')
plt.show()
plt.savefig('kdeplot.png')
```

```python
f = plt.figure()
iris.iloc[0:50].plot.area(alpha=0.4)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.savefig('areaplot.png')
```
