```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)

```

```python
fig = plt.figure(figsize=(16,16))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('A Ha Ha.png',dpi=200)
```

```python
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='sigma')
ax.plot(x,x**3,label='my bank acconnt')

ax.legend(loc=(0.1,0.1))
plt.savefig('bim',dpi=200)
```
