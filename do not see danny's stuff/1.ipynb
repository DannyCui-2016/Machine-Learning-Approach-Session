```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(iris)
plt.savefig('scatterplot.png')
```

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap="magma")
plt.savefig('magma.png')
```

```python
sns.swarmplot(x='species', y='sepal_length', data=iris, cmap='plasma')
plt.savefig('plasma.png')
```
