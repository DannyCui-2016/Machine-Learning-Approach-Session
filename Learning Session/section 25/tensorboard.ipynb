```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cancer_classification2.csv')
```

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
from datetime import datetime
datetime.now().strftime("%Y-%m-%d_%H%M")
```

```python
log_directory = "logs/fit/"
```

```python
board = TensorBoard(log_dir=log_directory, histogram_freq=1, write_graph=True, write_images=True, update_freq='epoch', profile_batch=2, embeddings_freq=0, embeddings_metadata=None)
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[board,early_stop])
```
