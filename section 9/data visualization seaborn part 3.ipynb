```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

```python
sns.heatmap(tc)
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.heatmap(fp, cmap="magma")
plt.savefig('heatmap.png')
```

```python
sns.heatmap(fp, cmap="plasma", linecolor='white', linewidth=1)
```

```python
sns.heatmap(fp, cmap="coolwarm", linecolor='white', linewidth=1)
plt.savefig('heatmap2.png')
```

```python
sns.clustermap(fp, cmap="cividis", standard_scale=1)
plt.savefig('clustermap.png')
```
