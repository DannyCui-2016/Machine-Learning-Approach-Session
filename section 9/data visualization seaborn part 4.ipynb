```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], palette='Set1')
plt.savefig('lmplot.png')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='smoker', palette='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time',hue='sex', palette='magma')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', palette='magma'
           , aspect=0.6, height=8)
plt.savefig('lmplot2.png')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',palette='cividis')
```
